```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys

%pylab inline
pylab.rcParams['figure.figsize'] = (20.0, 10.0)

%load_ext autoreload
%autoreload 2

sys.path.append('..')
```

```python
import postprocessing as pp
```

```python
res1 = pp.process_results_file('../results/results_p1.txt')
res2 = pp.process_results_file('../results/results_p2.txt')
res3 = pp.process_results_file('../results/results_p3.txt')
res4 = pp.process_results_file('../results/results_p4.txt')
```

```python
data = pd.concat([res1, res2, res3, res4])
data.index.name = 'algorithm'
data = data.reset_index()
data
```

## Use a table or chart to analyze the number of nodes expanded against number of actions in the domain

```python
sub = data[['algorithm', 'actions', 'expansions']]
sub
```

```python
sub_table = sub.pivot(index='actions', columns='algorithm', values='expansions')
sub_table = sub_table.sort_values(by=88, axis=1)
sub_table.to_csv('../results/actions_expansions.csv')
sub_table
```

```python
transpose = sub_table.T
transpose.to_csv('../results/actions_expansions2.csv')
transpose
```

```python
sub_table.mean(axis=1)
```

```python
_ = sub_table.plot(logy=True)
plt.ylabel('expansions')
plt.grid(True, 'minor', 'both')
_ = plt.title('Expansions vs actions for each algorithm')
```

## Use a table or chart to analyze the search time against the number of actions in the domain 

```python
sub = data[['algorithm', 'actions', 'time_s']]
sub.time_s = np.round(sub.time_s, 3)
sub_table = sub.pivot(index='actions', columns='algorithm', values='time_s')
sub_table = sub_table.sort_values(by=88, axis=1)
sub_table.to_csv('../results/actions_time.csv')
sub_table
```

```python
transpose = sub_table.T
transpose.to_csv('../results/actions_time2.csv')
transpose
```

```python
_ = sub_table.plot(logy=True)
plt.ylabel('time (s)')
plt.grid(True, 'minor', 'both')
_ = plt.title('Search time vs actions for each algorithm')
```

## Use a table or chart to analyze the length of the plans returned by each algorithm on all search problems 

```python
sub = data[['algorithm', 'actions', 'path_length']]
sub_table = sub.pivot(index='actions', columns='algorithm', values='path_length')
sub_table = sub_table.sort_values(by=88, axis=1)
sub_table = sub_table.reset_index(drop=True)
sub_table.index = range(1, 5)
sub_table.index.name = 'problem'
sub_table.to_csv('../results/actions_length.csv')
sub_table
```

```python
s_ids = {name: idx for name,idx in zip(res1.index, range(1,12))}
s_ids
```

```python
sub_table.columns.to_series()
```

```python
sub_table.columns = sub_table.columns.to_series().map(s_ids)
sub_table
```

```python
for i in range(1,5):
    plt.subplot(2,2,i)
    plt.scatter(sub_table.loc[i].index, sub_table.loc[i])
    plt.title('Problem {}'.format(i))
    plt.xlabel('algorithm')
    plt.ylabel('path length')
```
